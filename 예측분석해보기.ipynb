```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA

# 데이터 불러오기
monthly_charge = pd.read_csv('2년구별총충전량.csv', index_col=0, encoding='cp949')

# 인덱스를 날짜로 설정
date_index = pd.date_range(start='2020-01-01', periods=len(monthly_charge), freq='MS')
monthly_charge.index = date_index

# 구 리스트
districts = ['광산구', '남구', '동구', '북구', '서구']

# 예측 기간 설정
start_date = '2022-01-01'
end_date = '2023-10-01'
forecast_steps = 22

# X축 레이블 조정
xticks_dates = pd.date_range(start='2020-01-01', end=end_date, freq='MS')
xticks_labels = [f"{x.year}-{x.month:02d}" for x in xticks_dates]

# 시각화 준비
plt.figure(figsize=(12, 6))

# 각 구별로 ARIMA 모델로 월별 총 충전량 시계열 예측
for district in districts:
    train_data = monthly_charge.loc[:'2021-10', [district]]

    # ARIMA 모델 훈련
    arima_model = ARIMA(train_data, order=(1,1,1))
    arima_result = arima_model.fit()

    # 예측
    forecast = arima_result.get_forecast(steps=forecast_steps)
    forecast_conf_int = forecast.conf_int()

    # 예측 기간 날짜 생성
    forecast_dates = pd.date_range(start=start_date, periods=forecast_steps, freq='MS')

    # 시각화
    plt.plot(monthly_charge.index, monthly_charge[district], label=f'{district} Observed')
    plt.plot(forecast_dates, forecast.predicted_mean.values, label=f'{district} Forecast', linestyle='--')

plt.xticks(xticks_dates[::2], xticks_labels[::2], rotation=90)

plt.title('광주광역시 구별 월별 총 충전량 시계열 예측(ARIMA모델)')
plt.xlabel('기간')
plt.ylabel('총 충전량')
plt.legend()
plt.show()
```

```python
import numpy as np
from sklearn.metrics import mean_squared_error

# 각 구별로 RMSE 계산
for district in districts:
    train_data = monthly_charge.loc[:'2021-10', [district]]

    # ARIMA 모델 훈련
    arima_model = ARIMA(train_data, order=(1, 1, 1))
    arima_result = arima_model.fit()

    # 예측
    forecast = arima_result.get_forecast(steps=forecast_steps)

    # 실제값과 예측값
    actual_values = monthly_charge.loc['2021-01':'2021-10', [district]].values
    predicted_values = forecast.predicted_mean.values[:10]  # 2021년 1월부터 10월까지의 예측값

    # RMSE 계산
    rmse = np.sqrt(mean_squared_error(actual_values, predicted_values))
    print(f'{district} RMSE: {rmse}')
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

# 데이터 불러오기
monthly_charge = pd.read_csv('2년구별총충전량.csv', index_col=0, encoding='cp949')

# 인덱스를 날짜로 설정
date_index = pd.date_range(start='2020-01-01', periods=len(monthly_charge), freq='MS')
monthly_charge.index = date_index

# 누락된 값에 선형으로 보간된 값을 채워 넣기
monthly_charge = monthly_charge.interpolate()

# 구 리스트
districts = ['광산구', '남구', '동구', '북구', '서구']

# 시계열 데이터를 랜덤 포레스트에 사용하기 적합한 형태로 변환
def create_lagged_features(data, lag=1):
    lagged_data = data.shift(lag)
    return lagged_data

# 예측 기간 설정
start_date = '2022-01-01'
end_date = '2023-10-01'
forecast_steps = 20

plt.figure(figsize=(12, 6))

# 각 구별로 랜덤 포레스트 모델로 월별 총 충전량 시계열 예측
for district in districts:
    data = monthly_charge[[district]]
    lagged_data = create_lagged_features(data, lag=1).dropna()
    
    X = lagged_data.values
    y = data.loc[lagged_data.index].values.ravel()

    X_train, X_test = X[:-forecast_steps], X[-forecast_steps:]
    y_train, y_test = y[:-forecast_steps], y[-forecast_steps:]

    # 랜덤 포레스트 회귀 모델 훈련
    rf_model = RandomForestRegressor(n_estimators=100, random_state=42)
    rf_model.fit(X_train, y_train)

    # 예측
    y_pred = rf_model.predict(X_test)

    # 시각화
    plt.plot(monthly_charge.index, monthly_charge[district], label=f'{district} Observed')
    plt.plot(pd.date_range(start=start_date, periods=forecast_steps, freq='MS'), y_pred, label=f'{district} Forecast', linestyle='--')

plt.title('광주광역시 구별 월별 총 충전량 시계열 예측 (랜덤 포레스트)')
plt.xlabel('기간')
plt.ylabel('총 충전량')
plt.legend()
plt.show()
```

```python
from math import sqrt

# 각 구별로 랜덤 포레스트 모델로 월별 총 충전량 시계열 예측
for district in districts:
    data = monthly_charge[[district]]
    lagged_data = create_lagged_features(data, lag=1).dropna()
    
    X = lagged_data.values
    y = data.loc[lagged_data.index].values.ravel()

    X_train, X_test = X[:-forecast_steps], X[-forecast_steps:]
    y_train, y_test = y[:-forecast_steps], y[-forecast_steps:]

    # 랜덤 포레스트 회귀 모델 훈련
    rf_model = RandomForestRegressor(n_estimators=100, random_state=42)
    rf_model.fit(X_train, y_train)

    # 예측
    y_pred = rf_model.predict(X_test)

    # 정확도 측정 (RMSE)
    rmse = sqrt(mean_squared_error(y_test, y_pred))
    print(f'{district} RMSE: {rmse}')

    # 시각화
    plt.plot(monthly_charge.index, monthly_charge[district], label=f'{district} Observed')
    plt.plot(pd.date_range(start=start_date, periods=len(y_pred), freq='MS'), y_pred, label=f'{district} Forecast', linestyle='--')

plt.title('광주광역시 구별 월별 총 충전량 시계열 예측 (랜덤 포레스트)')
plt.xlabel('기간')
plt.ylabel('총 충전량')
plt.legend()
plt.show()
```
